**Question 1**

Given two linked list of the same size, the task is to create a new linked list using those linked lists. The condition is that the greater node among both linked list will be added to the new linked list.

</aside>

```python
class Solution:
    def addGreaterNode(self, l1, l2):
        new_l = [0]*len(l1)
        if l1.val < l2.val:
            new_l.val = l2.val
            new_l.next = self.addGreaterNode(l1.next, l2.next)
            return new_l
        else:
            new_l.val = l1.val
            new_l.next = self.addGreaterNode(l2.next, l1.next)  
            return new_l
```

#### TC: O(l1+l2)
    
#### SC: O(n)

**Question 2**

Write a function that takes a list sorted in non-decreasing order and deletes any duplicate nodes from the list. The list should only be traversed once.

For example if the linked list is 11->11->11->21->43->43->60 then removeDuplicates() should convert the list to 11->21->43->60.



```python
class Solution:
    def removeDuplicates(self, head):
        if not head:
            return head
        slow, fast = head, head.next
        prev = None
        curr = slow
        if prev:    # if it is a head Node
            head = curr.next    # skipping the first Node
            curr = head
        while fast:
            if slow.val == fast.val:    # If value matches
                slow = slow.next.next    # Node skipped 
                fast = slow.next
                if curr.val == slow.val:    # if the next node also same
                    curr.next = slow.next.next    # moving curr and fast
                    fast = curr.next    # moving fast
                    slow = curr    # bring slow equal
                else:
                    pass
            else:
                slow = slow.next    # moving ahead
                fast = fast.next
        return head
```

#### TC: O(n)
    
#### SC: O(1)

**Question 3**

Given a linked list of size **N**. The task is to reverse every **k** nodes (where k is an input to the function) in the linked list. If the number of nodes is not a multiple of *k* then left-out nodes, in the end, should be considered as a group and must be reversed (See Example 2 for clarification).



```python
class Solution:
    def reverse_K_Nodes(self, head, K):
        slow = fast = head
        while fast:
            if fast = K:    # while K value matched
                temp = fast
                slow.next = fast.val    # bring to the 1st position
                slow = temp
            fast = fast.next    # moving forward
```

#### TC: O(n)
    
#### SC: O(1)

**Question 4**

Given a linked list, write a function to reverse every alternate k nodes (where k is an input to the function) in an efficient way. Give the complexity of your algorithm.



```python
class Soution:
    def reverseAlternativeK(self, head, k):
        slow = fast = head
        for i in range(k):
            fast = fast.next
        while fast:
            if fast == k:
                temp1 = fast.val
                temp2 = fast.next
                fast.val = slow.val
                fast.next = slow.next
                slow.val = temp1
                slow.next = temp2
            else:
                fast = fast.next
                slow = slow.next
        return head
```

#### TC: O(n)
    
#### SC: O(1)

**Question 6**

Given two sorted linked lists consisting of **N** and **M** nodes respectively. The task is to merge both of the lists (in place) and return the head of the merged list.



```python
class Solution:
    def mergeTwoLists(self, N, M):
        if N is None:
            return M
        if M is None:
            return N
        if N.val < M.val:
            N.next = self.mergeTwoLists(N.next, M)
            return N
        else:
            M.next = self.mergeTwoLists(M.next, N)
            return M
```

#### TC: O(n)
    
#### SC: O(1)

**Question 7**

Given a **Doubly Linked List**, the task is to reverse the given Doubly Linked List.



```python
class Solution:
    def reverseDoublyLL(self, head):

        if head is None:
            return head
        curr = head
        prev = None
        while curr or prev:    # untill the last node
            temp = curr.next    # storing to a variable    
            curr.next = prev    # reverse
            prev = curr    # moving ahead
            curr = temp
        return head
```

#### TC: O(n)
    
#### SC: O(1)

**Question 8**

Given a doubly linked list and a position. The task is to delete a node from given position in a doubly linked list.



```python
class Solution:
    def deleteNode(self, head, x):
        slow, fast = head, head.next
        if fast:
            if fast == x:
                fast = fast.next.next
            fast = fast.next
        return head
```

#### TC: O(n)
    
#### SC: O(1)
